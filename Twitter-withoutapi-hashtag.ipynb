# Twitter extract without API for hashtag

## Import Libraries

```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import json
```

## Query strategy

```python
query = "(#ElCambioEnPrimera) until:2022-06-30 since:2022-05-09" # this form for to search a hashtag 
tweets = []
limit = 50000
```

## Bucle for to build a tweets list

### Bucle with conditional for limits

```python
for tweet in sntwitter.TwitterHashtagScraper(query).get_items():
    
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.id, tweet.date, tweet.url, tweet.user.username, tweet.sourceLabel, tweet.user.location, tweet.content, tweet.retweetedTweet, tweet.quotedTweet, tweet.inReplyToUser, tweet.mentionedUsers, tweet.likeCount, tweet.retweetCount,  tweet.quoteCount, tweet.replyCount, tweet.hashtags])
```

### Bucle for to extract all tweets

```python
cont = 0
for tweet in sntwitter.TwitterHashtagScraper(query).get_items():
    tweets.append([tweet.id, tweet.date, tweet.url, tweet.user.username, tweet.sourceLabel, tweet.user.location, tweet.content, tweet.retweetedTweet, tweet.quotedTweet, tweet.inReplyToUser, tweet.mentionedUsers, tweet.likeCount, tweet.retweetCount,  tweet.quoteCount, tweet.replyCount, tweet.hashtags])
    cont+=1
    print("Recorrido: ",cont)
```

## To build a dataframe

```python
df_quint = pd.DataFrame(tweets, columns=['ID', 'Date', 'TweetURL','User', 'Source', 'Location', 'Tweet', 'Retweeted_Tweet', 'Quoted_Tweet', 'In_Reply_to_User', 'Mentioned_Users', 'Likes_Count','Retweet_Count', 'Quote_Count', 'Reply_Count', 'Hashtags'])
```

## To export the dataframe in CSV file

```python
df_quint.to_csv("elcambioenprimera.csv")
```
